```python
import pandas as pd
from scipy import signal, stats
```

```python
df_facility = pd.read_csv('/home/is/shuntaro-o/dev/compare_population_and_tweet_number/data/social_corr/overpass/facility_numbers.csv',index_col=0)
#df_corr = pd.read_csv('/home/is/shuntaro-o/dev/compare_population_and_tweet_number/outputs/統計/ソーシャルセンサ性能.csv',index_col=0)
df_corr_work = pd.read_csv('/home/is/shuntaro-o/dev/compare_population_and_tweet_number/outputs/統計/ソーシャルセンサ性能平日.csv',index_col=0)
df_corr_holi = pd.read_csv('/home/is/shuntaro-o/dev/compare_population_and_tweet_number/outputs/統計/ソーシャルセンサ性能休日.csv',index_col=0)
df_kankou = pd.read_csv('/home/is/shuntaro-o/dev/compare_population_and_tweet_number/data/social_corr/Kankou/Kankou.csv',index_col=0)
df_Kyozyu= pd.read_csv('/home/is/shuntaro-o/dev/compare_population_and_tweet_number/data/social_corr/kyozyuu/Kyozyu.csv',index_col=0)
df_distance = pd.read_csv('/home/is/shuntaro-o/dev/compare_population_and_tweet_number/data/social_corr/distance/cal_distance.csv',index_col=0)
```

```python
df_distance = df_distance.iloc[:,8]
```

```python
df_merge = df_corr_work.merge(df_corr_holi,left_index=True,right_index=True)
```

```python
df_merge = df_merge.merge(df_facility,left_index=True,right_index=True)
```

```python
df_merge = df_merge.merge(df_kankou,left_index=True,right_index=True)
```

```python
df_merge = df_merge.merge(df_Kyozyu,left_index=True,right_index=True)
```

```python
df_merge = df_merge.merge(df_distance,left_index=True,right_index=True)
```

```python
r_r = []
for i in df_merge.columns:
    corr, pvalue = stats.pearsonr(df_merge['ソーシャルセンサ性能(平日)'], df_merge[i])
    r_r.append([corr, pvalue])
df_r = pd.DataFrame(r_r,index=df_merge.columns, columns=['相関係数','p値'])
```
