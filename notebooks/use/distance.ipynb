```python
import pandas as pd
from geopy.distance import geodesic
```

```python
df_latlon = pd.read_csv('/home/is/shuntaro-o/dev/compare_population_and_tweet_number/data/social_corr/distance/distance_origin.csv')
```

```python
df_latlon
```

```python
distance = []
for i ,item in df_latlon.iterrows():
    if item['駅緯度']==0:
        distance_m = 0
    else:
        point = (item['緯度'],item['経度'])
        station = (item['駅緯度'],item['駅経度'])
        distance_m = geodesic(point, station).m 
    distance.append(distance_m)
```

```python
df_latlon['駅距離'] = distance
```
