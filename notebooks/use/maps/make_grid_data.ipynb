```python
def aggregate_mesh(x, level):
    tmp = x.copy()
    code_len = [4, 6, 8, 9, 10][level-1]
    tmp["key"] = tmp["KEY_CODE"].str.slice(0, code_len)
    tmp = tmp[["key", "geometry"]]
    tmp = tmp.dissolve(by="key")
    return tmp
```

```python
Tokyo_grid_mesh3 = aggregate_mesh(Tokyo_grid, 3)
Tokyo_boxs = Tokyo_grid_mesh3.bounds
#Tokyo_boxs['key'] = Tokyo_grid_mesh3['key']
```
