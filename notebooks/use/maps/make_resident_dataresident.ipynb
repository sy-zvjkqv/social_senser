```python
import pandas as pd
```

```python
df_Tokyo = pd.read_csv('/home/is/shuntaro-o/dev/compare_population_and_tweet_number/data/grid_data/residence/5339.txt')
df_Kyoto= pd.read_csv('/home/is/shuntaro-o/dev/compare_population_and_tweet_number/data/grid_data/residence/5235.txt')

Tokyo_grid = pd.read_csv('/home/is/shuntaro-o/dev/compare_population_and_tweet_number/data/grid_data/SDDSWS5339/Tokyo_grid.csv')
Kyoto_grid = pd.read_csv('/home/is/shuntaro-o/dev/compare_population_and_tweet_number/data/grid_data/QDDSWQ5235/Kyoto_grid.csv')
```

```python
Kyoto_grid
```

```python
df_Tokyo = df_Tokyo[['KEY_CODE','T001100001']]
df_Kyoto = df_Kyoto[['KEY_CODE','T001100001']]
```

```python
df_Tokyo = df_Tokyo.rename(columns={'T001100001': 'residents'})
df_Tokyo = df_Tokyo.rename(columns={'KEY_CODE':'key'})

df_Kyoto = df_Kyoto.rename(columns={'T001100001': 'residents'})
df_Kyoto = df_Kyoto.rename(columns={'KEY_CODE':'key'})
```

```python
Tokyo_merge = pd.merge(Tokyo_grid,df_Tokyo, on='key', how='left')
```

```python
Kyoto_merge = pd.merge(Kyoto_grid,df_Kyoto, on='key', how='left')
```
