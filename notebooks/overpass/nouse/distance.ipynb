```python
import pandas as pd
from geopy.distance import geodesic
```

```python
df_latlon = pd.read_csv('/home/is/shuntaro-o/dev/compare_population_and_tweet_number/data/distance.csv')
```

```python
df_latlon
```

```python
df_latlon['経度'] = (df_latlon['西'] + df_latlon['東'])/2
df_latlon['緯度'] = (df_latlon['南'] + df_latlon['北'])/2
```

```python
df_latlon
```

```python
distance = []
for i ,item in df_latlon.iterrows():
    if item['駅緯度']==0:
        distance_m = 0
    else:
        point = (item['緯度'],item['経度'])
        station = (item['駅緯度'],item['駅経度'])
        distance_m = geodesic(point, station).m 
    distance.append(distance_m)
```

```python
df_latlon['駅距離'] = distance
```

```python
df_latlon.to_csv('/home/is/shuntaro-o/dev/compare_population_and_tweet_number/data/distance/ca;_distance.csv')
```
